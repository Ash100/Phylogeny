```python
#@title Visualize alignment
import numpy as np
import matplotlib.pyplot as plt
from Bio import AlignIO

# Load the alignment file
alignment_file = "/content/aligned_sequences.fasta"
alignment = AlignIO.read(alignment_file, "fasta")

# Convert the alignment into a matrix
alignment_matrix = np.array([list(record.seq) for record in alignment])

# Define a color map for amino acids
amino_acid_colors = {
    "A": [1, 0, 0],      # Alanine (Red)
    "C": [0, 1, 0],      # Cysteine (Green)
    "D": [0, 0, 1],      # Aspartic Acid (Blue)
    "E": [1, 1, 0],      # Glutamic Acid (Yellow)
    "F": [1, 0, 1],      # Phenylalanine (Magenta)
    "G": [0, 1, 1],      # Glycine (Cyan)
    "H": [0.5, 0, 0],    # Histidine (Dark Red)
    "I": [0, 0.5, 0],    # Isoleucine (Dark Green)
    "K": [0, 0, 0.5],    # Lysine (Dark Blue)
    "L": [0.5, 0.5, 0],  # Leucine (Olive)
    "M": [0.5, 0, 0.5],  # Methionine (Purple)
    "N": [0, 0.5, 0.5],  # Asparagine (Teal)
    "P": [1, 0.5, 0],    # Proline (Orange)
    "Q": [0.5, 1, 0],    # Glutamine (Lime)
    "R": [0, 0, 1],      # Arginine (Blue)
    "S": [1, 0, 0.5],    # Serine (Pink)
    "T": [0, 1, 0.5],    # Threonine (Light Green)
    "V": [0.5, 0.5, 0.5],# Valine (Gray)
    "W": [0.7, 0.3, 0],  # Tryptophan (Brown)
    "Y": [0.3, 0.7, 0],  # Tyrosine (Dark Green)
    "-": [0, 0, 0],      # Gap (Black)
}

# Map amino acids to colors
color_matrix = np.array([[amino_acid_colors.get(base, [0.5, 0.5, 0.5]) for base in row] for row in alignment_matrix])

# Calculate conservation scores (frequency of the most common amino acid at each position)
conservation_scores = []
for col in range(alignment.get_alignment_length()):
    column = [record.seq[col] for record in alignment]
    freqs = {aa: column.count(aa) / len(column) for aa in set(column)}
    conservation_scores.append(max(freqs.values()))

# Create a combined figure with heatmap and conservation panel
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 12), gridspec_kw={'height_ratios': [8, 1]}, constrained_layout=True)

# Plot the heatmap
heatmap = ax1.imshow(color_matrix, aspect="auto", interpolation="nearest")
ax1.set_title("Protein Sequence Alignment Heatmap", fontsize=16)
ax1.set_xlabel("Position in Alignment", fontsize=14)
ax1.set_ylabel("Sequences", fontsize=14)
fig.colorbar(heatmap, ax=ax1, label="Amino Acid Color Code")

# Plot the conservation panel
ax2.bar(range(len(conservation_scores)), conservation_scores, color='gray', width=1)
ax2.set_xlim(0, alignment.get_alignment_length())
ax2.set_ylim(0, 1)
ax2.set_xlabel("Position in Alignment", fontsize=14)
ax2.set_ylabel("Conservation", fontsize=14)

# Save the combined figure
plt.savefig("protein_alignment_heatmap_with_conservation.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#@title Amino Acid Variation Analysis
# - Calculates position-wise amino acid frequencies
# - Generates visualization using matplotlib

# %%
from Bio import AlignIO
import matplotlib.pyplot as plt
from collections import Counter

alignment = AlignIO.read("aligned_sequences.fasta", "fasta")

variation_data = []
for i in range(alignment.get_alignment_length()):
    column = alignment[:, i]
    counts = Counter(column)
    variation_data.append(counts)

# Plot position-wise variation
positions = list(range(1, len(variation_data)+1))
num_variants = [len(v) for v in variation_data]

plt.figure(figsize=(12, 6))
plt.bar(positions, num_variants)
plt.xlabel('Alignment Position')
plt.ylabel('Number of Variants')
plt.title('Amino Acid Variation Across Positions')
plt.show()
```

```python
from Bio import Phylo
import graphviz

# Load the Newick tree file
tree_file = "phylogenetic_tree.nwk"
tree = Phylo.read(tree_file, "newick")

# Convert the tree to a Graphviz graph
dot = graphviz.Digraph()

def add_edges(clade, parent=None):
    if clade.name:
        node_name = clade.name
    else:
        node_name = str(id(clade))  # Use unique ID for unnamed nodes
    dot.node(node_name)
    if parent:
        dot.edge(parent, node_name)
    for child in clade.clades:
        add_edges(child, node_name)

add_edges(tree.root)

# Render the graph
dot.render("phylogenetic_tree", format="png", cleanup=True)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from Bio import Phylo

def visualize_phylogenetic_network(
    tree_file="phylogenetic_tree.nwk",
    layout="spring",  # Options: "spring", "circular", "kamada_kawai", "spectral"
    node_size=500,
    node_color="lightblue",
    font_size=8,
    font_weight="bold",
    edge_color="gray",
    figsize=(15, 10),
    title="Phylogenetic Network",
    save_path=None  # Optional: Path to save the figure (e.g., "output.png")
):
    """
    Visualize a phylogenetic tree as a network using NetworkX and Matplotlib.

    Parameters:
        tree_file (str): Path to the Newick tree file.
        layout (str): Layout algorithm for the network. Options: "spring", "circular", "kamada_kawai", "spectral".
        node_size (int): Size of the nodes in the network.
        node_color (str): Color of the nodes.
        font_size (int): Font size for node labels.
        font_weight (str): Font weight for node labels (e.g., "normal", "bold").
        edge_color (str): Color of the edges.
        figsize (tuple): Figure size (width, height) in inches.
        title (str): Title of the plot.
        save_path (str): Optional path to save the figure (e.g., "output.png").
    """
    # Load the Newick tree file
    tree = Phylo.read(tree_file, "newick")

    # Create a graph
    G = nx.Graph()

    def add_edges(clade, parent=None):
        """Recursively add edges to the graph."""
        if clade.name:
            node_name = clade.name
        else:
            node_name = str(id(clade))  # Use unique ID for unnamed nodes
        G.add_node(node_name)
        if parent:
            G.add_edge(parent, node_name)
        for child in clade.clades:
            add_edges(child, node_name)

    add_edges(tree.root)

    # Select layout algorithm
    if layout == "spring":
        pos = nx.spring_layout(G)
    elif layout == "circular":
        pos = nx.circular_layout(G)
    elif layout == "kamada_kawai":
        pos = nx.kamada_kawai_layout(G)
    elif layout == "spectral":
        pos = nx.spectral_layout(G)
    else:
        raise ValueError(f"Unsupported layout: {layout}. Choose from 'spring', 'circular', 'kamada_kawai', 'spectral'.")

    # Draw the network
    plt.figure(figsize=figsize)
    nx.draw(
        G,
        pos,
        with_labels=True,
        node_size=node_size,
        node_color=node_color,
        font_size=font_size,
        font_weight=font_weight,
        edge_color=edge_color
    )
    plt.title(title)

    # Save the figure if save_path is provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches="tight")

    # Show the plot
    plt.show()
```

```python
# Example usage with save functionality
visualize_phylogenetic_network(
    tree_file="phylogenetic_tree.nwk",
    save_path="phylogenetic_network_default.png"  # Saves the figure to this file
)
```

```python
from Bio import Phylo
import networkx as nx
import matplotlib.pyplot as plt

# Load the Newick tree file
tree_file = "phylogenetic_tree.nwk"
tree = Phylo.read(tree_file, "newick")

# Convert the tree to a NetworkX graph
G = nx.Graph()

def add_edges(clade, parent=None):
    """Recursively add edges to the graph."""
    if clade.name:
        node_name = clade.name
    else:
        node_name = str(id(clade))  # Use unique ID for unnamed nodes
    if parent:
        G.add_edge(parent, node_name)
    for child in clade.clades:
        add_edges(child, node_name)

add_edges(tree.root)
```

```python
from Bio import AlignIO

# Load the alignment file
alignment_file = "aligned_sequences.fasta"
alignment = AlignIO.read(alignment_file, "fasta")

# Extract sequences as strings
sequences = {record.id: str(record.seq) for record in alignment}

# Function to calculate Hamming distance
def hamming_distance(seq1, seq2):
    """Calculate the Hamming distance between two sequences."""
    if len(seq1) != len(seq2):
        raise ValueError("Sequences must have the same length.")
    return sum(c1 != c2 for c1, c2 in zip(seq1, seq2))

# Compute pairwise distances for connected nodes only
edge_distances = {}
for u, v in G.edges():
    if u in sequences and v in sequences:
        distance = hamming_distance(sequences[u], sequences[v])
        edge_distances[(u, v)] = distance
```

```python
# Draw the network
plt.figure(figsize=(15, 10))
pos = nx.spring_layout(G)  # Layout algorithm

# Draw nodes and edges
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=500,
    node_color="lightblue",
    font_size=8,
    font_weight="bold",
    edge_color="gray"
)

# Add edge labels (mutation counts)
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=edge_distances,
    font_size=8,
    label_pos=0.5  # Position of the label along the edge
)

plt.title("Phylogenetic Tree with Mutation Counts")
plt.show()
```
